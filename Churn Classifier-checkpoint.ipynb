```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
```

```python
revenue = pd.read_excel('Revenue_Remakes_Post_Acquisition.xlsx', sheet_name = 'Revenue', index_col = 'Ref')
revenue
```

```python
remakes_2019 = pd.read_excel('Revenue_Remakes_Post_Acquisition.xlsx', sheet_name = 'Remake_2019', index_col = 'Ref')
remakes_2019
```

```python
remakes_2020 = pd.read_excel('Revenue_Remakes_Post_Acquisition.xlsx', sheet_name = 'Remake_2020', index_col = 'Ref')
remakes_2020
```

```python
remakes_2021 = pd.read_excel('Revenue_Remakes_Post_Acquisition.xlsx', sheet_name = 'Remake_2021_YTD', index_col = 'Ref')
remakes_2021
```

```python
w_2019_remake = pd.merge(left = revenue, right = remakes_2019, how = 'left', left_index = True, right_index = True)
w_2020_remake = pd.merge(left = w_2019_remake, right = remakes_2020, how = 'left', left_index = True, right_index = True)
post_acq_sales = pd.merge(left = w_2020_remake, right = remakes_2021, how = 'left', left_index = True, right_index = True)

post_acq_sales
```

```python
# Reduces customers who churned prior to the acquisition phase (eg. were customers before July 2019)
pre_acq = post_acq_sales[(post_acq_sales['perc_remake_2019'].isnull()) &(post_acq_sales['perc_remake_2020'].isnull()) & (post_acq_sales['perc_remake_2021'].isnull())]
first_50 = pre_acq.iloc[:50]

post_acq = post_acq_sales.dropna(subset = ['perc_remake_2019','perc_remake_2020','perc_remake_2021'], how = 'all')

post_and_pre = pd.concat([first_50, post_acq])
post_and_pre
```

```python
# Handle missing values

# Method 1: 3 more features, each corresponding to NaN for the year
post_acq_sales1 = post_and_pre.copy()
post_acq_sales1['is_2019_remake_nan'] = post_acq_sales1['perc_remake_2019'].isna()
post_acq_sales1['is_2020_remake_nan'] = post_acq_sales1['perc_remake_2020'].isna()
post_acq_sales1['is_2021_remake_nan'] = post_acq_sales1['perc_remake_2021'].isna()
post_acq_sales1 = post_acq_sales1.fillna(0)

# Select columns of interest
features = ['7/1/2019', '8/1/2019',
       '9/1/2019', '10/1/2019', '11/1/2019', '1/1/2020', '2/1/2020',
       '3/1/2020', '4/1/2020', '5/1/2020', '6/1/2020', '7/1/2020', '8/1/2020',
       '9/1/2020', '10/1/2020', '11/1/2020', '12/1/2020', '1/1/2021',
       '2/1/2021', '3/1/2021', '4/1/2021', '5/1/2021', '6/1/2021', 'perc_remake_2019', 'perc_remake_2020',
       'perc_remake_2021', 'is_2019_remake_nan', 'is_2020_remake_nan',
       'is_2021_remake_nan']
output = 'Label 07-09'

X1 = post_acq_sales1[features]
y1 = post_acq_sales1[output]

# Method 2: Impute NaN with -1
post_acq_sales2 = post_and_pre.copy()
post_acq_sales2 = post_acq_sales2.fillna(-10)

# Select columns of interest
features = ['7/1/2019', '8/1/2019',
       '9/1/2019', '10/1/2019', '11/1/2019', '1/1/2020', '2/1/2020',
       '3/1/2020', '4/1/2020', '5/1/2020', '6/1/2020', '7/1/2020', '8/1/2020',
       '9/1/2020', '10/1/2020', '11/1/2020', '12/1/2020', '1/1/2021',
       '2/1/2021', '3/1/2021', '4/1/2021', '5/1/2021', '6/1/2021', 'perc_remake_2019', 'perc_remake_2020',
       'perc_remake_2021']
output = 'Label 07-09'

X2 = post_acq_sales2[features]
y2 = post_acq_sales2[output]

```

```python
X1
```

```python
# Divide the data into training and testing data. Following the convention of 70% training data, 30% testing data
X_train, X_test, y_train, y_test = train_test_split(X1, y1, train_size = 0.70)
```

```python
dtree_parameters = {
    'max_depth': [2, 3, 4, 5, 7, 9, 10, 15, None],
    'min_samples_split': [2, 3, 5, 7, 10, 13, 15, 20],
    'max_leaf_nodes': [2, 3, 5, 7, None]
}
```

```python
dtree = GridSearchCV(DecisionTreeClassifier(), dtree_parameters, cv = 5)
```

```python
dtree.fit(X_train, y_train)
```

```python
dtree.best_params_
```

```python
# Utilizing Decision Tree
y_predictions = dtree.predict(X_test) # testing
```

```python
metrics.precision_score(y_test, y_predictions)
```

```python
metrics.recall_score(y_test, y_predictions)
```

```python
metrics.f1_score(y_test, y_predictions)
```
